```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
# making the 1st column, index
df_product_a = pd.read_csv("product_a.csv", index_col=0)
print(df_product_a)
```

```python
# Convert date_w field to a suitable datetime data type
df_product_a['date_w'] = pd.to_datetime(df_product_a.date_w)
df_product_a['date_w'] = df_product_a['date_w'].dt.strftime('%m/%d/%Y')
print(df_product_a)
```

```python
# Correct the values of the year column.
df_product_a['date_w'] = pd.to_datetime(df_product_a.date_w)
df_product_a['year'] = df_product_a['date_w'].dt.year
print(df_product_a)
```

```python
# Descriptive Statistics 
df_stats=df_product_a.describe(include='all').transpose()
print(df_stats)
```

```python
# Pearson Correlation

pearsoncorr = df_product_a.corr(method='pearson')
pearsoncorr
```

```python
# Pearson Corr Heat map
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```

```python
spearmancorr = df_product_a.corr(method='spearman')
spearmancorr
```

```python
# Pearson Corr Heat map
sb.heatmap(spearmancorr, 
            xticklabels=spearmancorr.columns,
            yticklabels=spearmancorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```

```python
# Create the default pairplot
sb.pairplot(df_product_a)
```

```python
# Plotly
fig = go.Figure([go.Scatter(x=df_product_a['date_w'], y=df_product_a['price'])])
fig.show()
```

```python
# Time series, Monthly for records in Albany of Type 'A' and the price more than 1.5
df = df_product_a.loc[df_product_a['location'] == 'Albany']
df=df[df.type == 'A']
df=df[df.price > 1.5]
fig = go.Figure([go.Scatter(x=df['date_w'], y=df['price'])])
fig.show()
```
